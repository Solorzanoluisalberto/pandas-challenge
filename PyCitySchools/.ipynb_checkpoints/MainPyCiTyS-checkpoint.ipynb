```python
# main script to run for analysis
import collections
import csv
import pandas as pd
import numpy as np
```

```python
#files with data for analysis
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
schools = pd.read_csv(school_data)
students = pd.read_csv(student_data)

#print(schools)
#print(students)
```

```python
# Combine the data into a single dataset.  
All_school = pd.merge(schools, students, how="left", on=["school_name", "school_name"])
All_school
```

```python
# Create a high level snapshot (in table form) of the district's key metrics, including...
# Calculate the total number of schools
All_schools = set(schools["school_name"]) # unique school
tot_schools = len(All_schools) # number of school
# Calculate the total number of students
tot_Students = len(set(students["Student ID"]))
#print(tot_Students)
# Calculate the total budget
tot_budget = "${:,.0f}".format(sum(schools["budget"]))  # {:,}".format(100000) = 100,000
print(tot_budget)
# Calculate the average math score
average_math_score = "{:,.2f}".format(students["math_score"].mean())
print(average_math_score)
# Calculate the average reading score
#average_reading_score = round(students["reading_score"].mean(),2)
average_reading_score = "{:,.2f}".format(students["reading_score"].mean())
print(average_reading_score)
#Calculate the percentage of students with a passing math score (70 or greater)
stu70G = students[students["math_score"] >= 70] 
Stu_math_70G = stu70G['Student ID'].count() # all students passing math 70 or greater 
print(Stu_math_70G)
#Stu_math_70G_Per = "{:.3%}".format((Stu_math_70G/tot_Students)) # % students passing math 70 or greater (error operate str)
Stu_math_70G_Per = (Stu_math_70G/tot_Students) # remains to be expressed as a percentage
print("{:.4%}".format(Stu_math_70G_Per))
#Calculate the percentage of students with a passing reading score (70 or greater)
Stu_read_70G = students[students["reading_score"]>=70] # reading score (70 or greater)
Stu_reading_70G = Stu_read_70G["Student ID"].count() # number of student reading score (70 or greater)
#Stu_reading_70G_per = "{:.3%}".format((Stu_reading_70G/tot_Students))# % students passing read 70 or greater (error operate str)
Stu_reading_70G_per = (Stu_reading_70G/tot_Students) # remains to be expressed as a percentage
print("{:.4%}".format(Stu_reading_70G_per))
#Calculate the percentage of students who passed math and reading (% Overall Passing)
Overall_Passing = (Stu_math_70G_Per + Stu_reading_70G_per)/2
print("{:.4%}".format(Overall_Passing))
 
```

```python
#Create a dataframe to hold the above results

District_Summary_snapshot  = pd.DataFrame({"Total Schools":[tot_schools], "Total Students":["{:,.0f}".format(tot_Students)],
                                 "Total Budget":[tot_budget], "Average Math Score":[average_math_score],
                                 "Average Reading Score":[average_reading_score],
                                  "% Passing Math":["{:.4%}".format(Stu_math_70G_Per)],
                                 "% Passing Reading":["{:.4%}".format(Stu_reading_70G_per)], 
                                  "% Overall Passing":["{:.4%}".format(Overall_Passing)]})
District_Summary_snapshot
```

```python
# Create an overview table that summarizes key metrics about each school, including:
# School Type
School_Type = schools.set_index(["school_name"])["type"]  # set school name + school type
#print(School_Type)
```

```python
# Total Students
School_Total_Students = All_school["school_name"].value_counts()
#print(School_Total_Students)
```

```python
# Total School Budget
# Using GroupBy in order to separate the data into fields according to "School" values
Group_School_Budget = schools.groupby(['school_name'])
Total_School_Budget = Group_School_Budget['budget'].sum()
print(Total_School_Budget)
```

```python
# Per Student Budget 
Per_Student_Budget = Total_School_Budget/School_Total_Students
print(Per_Student_Budget)
```

```python
# Average Math Score
# Using GroupBy in order to separate the data into fields according to "school_name" values math_score
Average_Math_Score = All_school.groupby(["school_name"]).mean()["math_score"]
print(Average_Math_Score)
```

```python
# Average Reading Score
Average_Reading_Score = All_school.groupby(["school_name"]).mean()["reading_score"]

print(Average_Reading_Score)
```

```python
# % Passing Math (The percentage of students that passed math.)
Passing_Math = All_school[(All_school["math_score"]>=70)]
passing_math_x_school = Passing_Math.groupby(["school_name"]).count()["Student ID"]
percentage_Passing_Math_x_School = passing_math_x_school/School_Total_Students  # remains to be expressed as a percentage
print(percentage_Passing_Math_x_School)
```

```python
# % Passing Reading (The percentage of students that passed reading.)
Passing_Reading = All_school[(All_school["reading_score"]>=70)]
passing_Reading_x_school = Passing_Reading.groupby(["school_name"]).count()["Student ID"]
percentage_Passing_Reading_x_School = passing_Reading_x_school/School_Total_Students
print(percentage_Passing_Reading_x_School)
```

```python
# % Overall Passing (The percentage of students that passed math **and** reading.)
overall_passing = (percentage_Passing_Math_x_School + percentage_Passing_Reading_x_School)/2
print(overall_passing)
```

```python
# Create an overview table that summarizes key metrics about each school
School_Summary_overview = pd.DataFrame({"School Type":School_Type, "Total Students":School_Total_Students,
                                      "Total School Budget":Total_School_Budget, "Per Student Budget":Per_Student_Budget,
                                      "Average Math Score":Average_Math_Score,
                                      "Average Reading Score":Average_Reading_Score,
                                      "% Passing Math":round(percentage_Passing_Math_x_School*100,2),
                                      "% Passing Reading":round(percentage_Passing_Reading_x_School*100,2),
                                      "% Overall Passing":round(overall_passing*100,2)})
School_Summary_overview
```

```python
#Top Performing Schools (By % Overall Passing)
# To sort from highest to lowest, ascending=False must be passed in
top_Performing_Overall_Passing = School_Summary_overview.sort_values("% Overall Passing", ascending=False)
top_Performing_Overall_Passing.head(5)
```

```python
# Bottom Performing Schools (By % Overall Passing)
Bottom_Performing_Overall_Passing = School_Summary_overview.sort_values("% Overall Passing", ascending=True)
Bottom_Performing_Overall_Passing.head(5)
```

```python
# Math Scores by Grade\*\*
# Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement.
math_scores_by_grade = round(All_school.groupby(["grade"]).mean()["math_score"],2)
#product = math_scores_by_grade['grade'].values.tolist()
#df = pd.DataFrame(products, columns= ['Product', 'Price'])

#math_scores_by_gradew = All_school['grade'].values.tolist()
unique_grade = All_school.grade.unique() # get unique grade
# print(match_uniq_grade) ==> ['9th' '12th' '11th' '10th'] list values
#for x in range(0,len(match_uniq_grade)):
#    print(match_uniq_grade[x])
    
graders_9th = All_school[(All_school["grade"] == unique_grade[0])]
graders_10th = All_school[(All_school["grade"] == unique_grade[1])]
graders_11th = All_school[(All_school["grade"] == unique_grade[2])]
graders_12th = All_school[(All_school["grade"] == unique_grade[3])]

#.group by column math_score
graders_9th_scores = graders_9th.groupby(["school_name"]).mean()["math_score"]
graders_10th_scores = graders_10th.groupby(["school_name"]).mean()["math_score"]
graders_11th_scores = graders_11th.groupby(["school_name"]).mean()["math_score"]
graders_12th_scores = graders_12th.groupby(["school_name"]).mean()["math_score"]
Math_Scores_by_Grade= pd.DataFrame({"9th":round(graders_9th_scores,4),
                                "10th":round(graders_10th_scores,4),
                                "11th":round(graders_11th_scores,4),
                                "12th":round(graders_12th_scores,4)})

Math_Scores_by_Grade
```

```python
### Reading Scores by Grade
# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#.group by column reading_score
graders_9th_scores = graders_9th.groupby(["school_name"]).mean()["reading_score"]
graders_10th_scores = graders_10th.groupby(["school_name"]).mean()["reading_score"]
graders_11th_scores = graders_11th.groupby(["school_name"]).mean()["reading_score"]
graders_12th_scores = graders_12th.groupby(["school_name"]).mean()["reading_score"]
Reading_Scores_by_Grade= pd.DataFrame({"9th":round(graders_9th_scores,4),
                                "10th":round(graders_10th_scores,4),
                                "11th":round(graders_11th_scores,4),
                                "12th":round(graders_12th_scores,4)})
Reading_Scores_by_Grade
```

```python
#* Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
#Use 4 reasonable bins to (from PyCitySchools_starter)
Bins = [0, 585, 630, 645, 680]
# Create the names for the five bins (Activities/03-Ins_Binning)
group_names = ["< $585", "$585-630", "$630-645", "$645-680"]
School_Summary_overview["Spending Ranges (Per Student)"] = pd.cut(School_Summary_overview["Per Student Budget"], Bins, labels=group_names, include_lowest=True)
School_Summary_overview

```

```python
# Create a GroupBy object based upon "Spending Ranges (Per Student)"
#Spending_Ranges_group = School_Summary_overview.groupby("Spending Ranges (Per Student)")

Average_Math_Spending = School_Summary_overview.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
Average_Reading_Spending = School_Summary_overview.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
percentage_Passing_Math_Spending = School_Summary_overview.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
percentage_Passing_Reading_Spending = School_Summary_overview.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_Spending = School_Summary_overview.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

Spending_Ranges_group = pd.DataFrame({"Average Math Score":round(Average_Math_Spending,2),
                                      "Average Reading Score":round(Average_Reading_Spending,2),
                                      "% Passing Math":round(percentage_Passing_Math_Spending,2),
                                      "% Passing Reading":round(percentage_Passing_Reading_Spending,2),
                                      "% Overall Passing":round(overall_passing_Spending,2)})
Spending_Ranges_group
```

```python
### Scores by School Size
# Repeat the above breakdown, but this time group schools based on a reasonable approximation of school size (Small, Medium, Large).
School_Summary_Size = School_Summary_overview
del School_Summary_Size["Spending Ranges (Per Student)"]
Bins = [0, 1000, 2000, 5000]
# Create the names for the five bins (Activities/03-Ins_Binning)
group_names = ["Small (< 1000)", "Medium (1000-2000)", "Large (2000-5000)"]

School_Summary_Size["School Size"] = pd.cut(School_Summary_Size["Total Students"], Bins, labels=group_names, include_lowest=True)
 # Remove the 'Spending Ranges (Per Student)' column
#School_Summary_Size
Average_Math_Size = School_Summary_Size.groupby(["School Size"]).mean()["Average Math Score"]
Average_Reading_Size = School_Summary_Size.groupby(["School Size"]).mean()["Average Reading Score"]
percentage_Passing_Math_Size = School_Summary_Size.groupby(["School Size"]).mean()["% Passing Math"]
percentage_Passing_Reading_Size = School_Summary_Size.groupby(["School Size"]).mean()["% Passing Reading"]
overall_passing_Size = School_Summary_Size.groupby(["School Size"]).mean()["% Overall Passing"]

Size_Ranges_group = pd.DataFrame({"Average Math Score":round(Average_Math_Size,2),
                                      "Average Reading Score":round(Average_Reading_Size,2),
                                      "% Passing Math":round(percentage_Passing_Math_Size,2),
                                      "% Passing Reading":round(percentage_Passing_Reading_Size,2),
                                      "% Overall Passing":round(overall_passing_Size,2)})
Size_Ranges_group
```
